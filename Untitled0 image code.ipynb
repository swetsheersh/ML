```python
import numpy as np
from PIL import Image
import glob
image_list = []
for filename in glob.glob('./rice-image-dataset/Rice_Image_Dataset/Basmati/*.jpg'):
  #print(filename)
  im=np.array(Image.open(filename))
  image_list.append(im)

import pandas as pd
k=0
d=len(image_list)
(a,b,c)=map(int,image_list[0].shape)
new_df1=pd.DataFrame(np.zeros((1000,a*b*c),dtype=int))
for x in range(1000):
  new=image_list[x].reshape(a*b*c)
  new_df1.loc[k]=new
  k+=1

```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50,whiten=True,random_state=0,svd_solver='full')
pca.fit(new_df1)
x_pca1=pd.DataFrame(pca.transform(new_df1))
x_pca1['label']="Basmati"
```

```python
frames = [x_pca, x_pca1]
  
result = pd.concat(frames)
result
```

```python
import numpy as np
from PIL import Image
import glob
image_list = []
for filename in glob.glob('./rice-image-dataset/Rice_Image_Dataset/Ipsala/*.jpg'):
  #print(filename)
  im=np.array(Image.open(filename))
  image_list.append(im)

import pandas as pd
k=0
d=len(image_list)
(a,b,c)=map(int,image_list[0].shape)
new_df2=pd.DataFrame(np.zeros((1000,a*b*c),dtype=int))
for x in range(1000):
  new=image_list[x].reshape(a*b*c)
  new_df2.loc[k]=new
  k+=1

from sklearn.decomposition import PCA
pca=PCA(n_components=50,whiten=True,random_state=0,svd_solver='full')
pca.fit(new_df2)
x_pca2=pd.DataFrame(pca.transform(new_df2))
x_pca2['label']="Ipsala"
```
